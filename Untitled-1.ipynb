Import Required Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib 
from matplotlib import pyplot as plt
```

Load The Dataset

```python
# Load dataset 
file = r"E:/Datasets.csv/online_retail.csv"
ds = pd.read_csv(file)
print(ds.head(2))
```

# Preproccesing to clean and validate the data

```python
# Geting some insights
print("Data Info",ds.info())
```

```python
print("Data Count\n",ds.count())
```

```python
# Dropping missing values
#print(ds.dropna())
ds = ds.dropna()
```

```python
# Checking missing values for over all dataset
print("Missing Values for \n",ds.isnull().sum())
```

```python
# Drop irrelevant columns
ds = ds.drop(columns=['CustomerID'])
print(ds.head(2))
```

# Sales Trend Analysis

```python
# Extract yearmonth from dataset
ds['InvoiceDate'] = pd.to_datetime(ds['InvoiceDate'])
```

```python
# Seprate YearMonth
ds['YearMonth'] = ds['InvoiceDate'].dt.to_period('M')
print(ds.head(5))
```

```python
# Group by YearMonth and sum quantity
monthly_sales = ds.groupby('YearMonth')['Quantity'].sum().reset_index()
print(monthly_sales)
```

```python
# Checking datatype for new variable
print(monthly_sales.dtypes)
```

```python
# Convert YearMonth from period to datetime
monthly_sales['YearMonth'] = monthly_sales['YearMonth'].dt.to_timestamp()
```

```python
# Plotting monthly sales
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='YearMonth', y='Quantity')
plt.title('Monthly Sales Quantity Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Quantity')
plt.show()
```

Analyze Top Selling Products

```python
# Product Analysis
product_sales = ds.groupby(['StockCode', 'Description'])['Quantity'].sum().reset_index()
best_selling_products = product_sales.sort_values(by='Quantity', ascending=False).head(10)
```

```python
#Analyze top selling products
# Plot data usnig barplot  
plt.figure(figsize=(12, 6))
sns.barplot(data=best_selling_products, x='Quantity', y='Description')
plt.title('Top 10 Best-Selling Products')
plt.xlabel('Total Quantity Sold')
plt.ylabel('Product Description')
plt.xticks(rotation=45)
plt.show()
```
